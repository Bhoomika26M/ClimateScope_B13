```python
import pandas as pd

df = pd.read_csv(r"C:\Users\mailv\Desktop\projects\infosys\climatescope\cleaned_weather.csv")
```

```python
#replacing outliers
import numpy as np

df.replace(-9999, np.nan, inplace=True)
df.isnull().sum()
```

```python
print("Max wind_kph:", df["wind_kph"].max())
print("Max pressure_mb:", df["pressure_mb"].max())
print("Min PM10:", df["air_quality_pm10"].min())
```

```python
df = df[df["wind_kph"] < 400]
df = df[(df["pressure_mb"] > 800) & (df["pressure_mb"] < 1100)]
df = df[df["air_quality_pm10"] >= 0]
print("Rows after cleaning:", len(df))
```

```python
df = df.dropna()
```

```python
print(len(df))
```

```python
df.to_csv("final_cleaned_weather.csv", index=False)
```
