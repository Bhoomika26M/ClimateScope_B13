```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
df=pd.read_csv('cleaned_GlobalweatherRepository.csv')
```

```python
date=pd.to_datetime(df['last_updated_epoch'], unit='s')
yearly_temp = df.groupby(date.dt.year)['temperature_celsius'].mean()

plt.figure()
plt.plot(yearly_temp.index, yearly_temp.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.title('Year-wise Average Temperature')
plt.grid(True)
plt.show()
```

```python
date=pd.to_datetime(df['last_updated_epoch'], unit='s')
monthly_temp = df.groupby(date.dt.month)['temperature_celsius'].mean()

plt.figure()
plt.plot(monthly_temp.index, monthly_temp.values, marker='o')
plt.xlabel('Month')
plt.ylabel('Average Temperature')
plt.title('Month-wise Temperature Pattern')
plt.grid(True)
plt.show()
```

```python
date=pd.to_datetime(df['last_updated_epoch'], unit='s')
season=((date.dt.month%12 + 3)//3).map({1:'Spring', 2:'Summer', 3:'Autumn', 4:'Winter'})    
season_temp = df.groupby(season)['temperature_celsius'].mean()
plt.figure()

plt.bar(season_temp.index, season_temp.values)
plt.xlabel('Season')
plt.ylabel('Avg Temperature')
plt.title('Season-wise Temperature')
plt.show()
```

```python
date=pd.to_datetime(df['last_updated_epoch'], unit='s')


yearly_rain = df.groupby(date.dt.year)['precip_mm'].sum()

plt.figure()
plt.bar(yearly_rain.index, yearly_rain.values)
plt.xlabel('Year')
plt.ylabel('Total Rainfall')
plt.title('Year-wise Rainfall')
plt.show()
```

```python
plt.figure()
frequency=df['temperature_celsius'].value_counts().sort_index()
plt.hist(df['temperature_celsius'], bins=30)
plt.xlabel('Temperature (°C)')
plt.ylabel('frequency')
plt.title('Temperature Distribution')
plt.show()
```

```python
plt.figure()
date=pd.to_datetime(df['last_updated_epoch'], unit='s')

frequency=df['precip_mm'].value_counts().sort_index()
monthly_rain = df.groupby(date.dt.month)['precip_mm'].sum()
plt.bar(monthly_rain.index, monthly_rain.values)
plt.ylabel('Precipitation (mm)')
plt.xlabel('frequency')
plt.title('Rainfall Distribution')
plt.show()
```

```python
plt.figure()
plt.boxplot(df['temperature_celsius'], vert=False)
plt.xlabel('Temperature (°C)')
plt.title('Temperature Box Plot')
plt.show()
```

```python
df['temp_rolling_7'] = df['temperature_celsius'].rolling(7).mean()
date=pd.to_datetime(df['last_updated_epoch'], unit='s')
plt.figure()
plt.plot(date, df['temperature_celsius'], label='Actual')
plt.plot(date, df['temp_rolling_7'], label='7-day Rolling Avg')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Trend with Rolling Average')
plt.legend()
plt.show()
```

```python
rain_counts = (df['precip_mm'] > 0).value_counts()

plt.figure()
plt.pie(
    rain_counts,
    labels=['Rainy Days', 'No Rain Days'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Rain vs No Rain Days')
plt.show()
```

```python
threshold = df['temperature_celsius'].quantile(0.95)
heatwave = df[df['temperature_celsius'] > threshold]

plt.figure()
plt.scatter(df['last_updated'], df['temperature_celsius'], label='Normal')
plt.scatter(heatwave['last_updated'], heatwave['temperature_celsius'], label='Heatwave')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.title('Heatwave Detection')
plt.legend()
plt.show()
```
