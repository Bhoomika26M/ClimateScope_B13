## Importing the data 


```python
import numpy as np
import pandas as pd

raw_df=pd.read_csv("GlobalWeatherRepository.csv")
df=raw_df.copy()
```

### Check whather the data is fetched properly

```python
print(df)
```

### Making all the column names in small caps


```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace(".", "_", regex=False)
      .str.replace("-", "_", regex=False)
)
print(df.columns)
```

### Do a through check about the dataset  

```python
df.head()
df.shape
df.info()
```

### Handling the missing values and it's repersentation

```python
df.replace(
    ["NA", "null", "None", "", " "],
    pd.NA,
    inplace=True
)
```

### Converting the numerical datatime repersentation to real datatime representation

```python
df["last_updated_epoch"] = pd.to_numeric(
    df["last_updated_epoch"], errors="coerce"
)
# coerve to datetime, errors will be set to NaT

df["last_updated_dt"] = pd.to_datetime(
    df["last_updated_epoch"], unit="s", errors="coerce"
)
df["last_updated"] = pd.to_datetime(
    df["last_updated"], errors="coerce"
)
print(df[["last_updated_dt", "last_updated"]].head())
```

### Convert Numeric Columns Safely

```python
numeric_cols = [
    "latitude","longitude",
    "temperature_celsius","temperature_fahrenheit",
    "feels_like_celsius","feels_like_fahrenheit",
    "humidity","cloud","uv_index",
    "wind_mph","wind_kph","wind_degree",
    "gust_mph","gust_kph",
    "pressure_mb","pressure_in",
    "precip_mm","precip_in",
    "visibility_km","visibility_miles",
    "moon_illumination",
    "air_quality_carbon_monoxide",
    "air_quality_ozone",
    "air_quality_nitrogen_dioxide",
    "air_quality_sulphur_dioxide",
    "air_quality_pm2_5",
    "air_quality_pm10",
    "air_quality_us_epa_index",
    "air_quality_gb_defra_index"
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
        
print(df[numeric_cols].info())
```

### Drop Redundant Unit Columns Senior Decision

```python
cols_to_drop = [
    "temperature_fahrenheit",
    "feels_like_fahrenheit",
    "wind_mph",
    "gust_mph",
    "pressure_in",
    "precip_in",
    "visibility_miles"
]

df.drop(columns=[c for c in cols_to_drop if c in df.columns], inplace=True)
print(df.columns)
```

### Validate Physical Ranges and handles outliers

```python
df = df[
    (df["humidity"].between(0, 100, inclusive="both")) &
    (df["cloud"].between(0, 100, inclusive="both")) &
    (df["uv_index"] >= 0) &
    (df["wind_degree"].between(0, 360, inclusive="both")) &
    (df["latitude"].between(-90, 90, inclusive="both")) &
    (df["longitude"].between(-180, 180, inclusive="both"))
]
df.info()
```

### Hnadling the text data and formating that in a standard format 

```python
text_cols = ["country", "location_name", "condition_text", "wind_direction"]

for col in text_cols:
    if col in df.columns:
        df[col] = df[col].str.lower().str.strip()
print(df[text_cols].head())
```

### Handle Duplicates (Time-Series Safe)

```python
df.drop_duplicates(
    subset=["location_name", "last_updated_dt"],
    inplace=True
)
print(df.info())
```

### Final Validation

```python
df.info()
df.describe()
df.isna().sum()
```
