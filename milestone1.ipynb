```python
#removing columns that are in the same just in different units(e.g. temperature in Celsius and Fahrenheit)
columns_to_drop = [
    "temperature_fahrenheit",
    "wind_mph",
    "pressure_in",
    "precip_in",
    "visibility_miles",
    "gust_mph",
    "feels_like_fahrenheit"
]

df = df.drop(columns=columns_to_drop)
```

```python
#converting to date format
df["last_updated"] = pd.to_datetime(df["last_updated"])
df["year"] = df["last_updated"].dt.year
df["month"] = df["last_updated"].dt.month
df["day"] = df["last_updated"].dt.day
```

```python
#standardizing column names
df.columns = (
    df.columns
      .str.lower()
      .str.strip()
      .str.replace(" ", "_")
      .str.replace("-", "_")
      .str.replace(".", "", regex=False)
)
```

```python
df.describe()
```

```python
#defining seasons
def get_season(row):
    month = row["month"]
    lat = row["latitude"]
    
    # Northern Hemisphere
    if lat >= 0:
        if month in [12,1,2]:
            return "Winter"
        elif month in [3,4,5]:
            return "Spring"
        elif month in [6,7,8]:
            return "Summer"
        else:
            return "Autumn"
    
    # Southern Hemisphere (reverse)
    else:
        if month in [12,1,2]:
            return "Summer"
        elif month in [3,4,5]:
            return "Autumn"
        elif month in [6,7,8]:
            return "Winter"
        else:
            return "Spring"

df["season"] = df.apply(get_season, axis=1)
```

```python
#checking for outliers
print("Temperature (°C):")
print("Min:", df["temperature_celsius"].min())
print("Max:", df["temperature_celsius"].max())
print()

print("Humidity (%):")
print("Min:", df["humidity"].min())
print("Max:", df["humidity"].max())
print()

print("UV Index:")
print("Min:", df["uv_index"].min())
print("Max:", df["uv_index"].max())
print()

print("Precipitation (mm):")
print("Min:", df["precip_mm"].min())
print("Max:", df["precip_mm"].max())
```

```python
#dataset coverage
print("Start date:", df["last_updated"].min())
print("End date:", df["last_updated"].max())
print("Unique days:", df["last_updated"].dt.date.nunique())
```

```python
monthly_summary = (
    df.groupby(["country", "year", "month"])
      .agg({
          "temperature_celsius": "mean",
          "humidity": "mean",
          "precip_mm": "sum",
          "wind_kph": "mean"
      })
      .reset_index()
)

monthly_summary.to_parquet("monthly_summary.parquet", index=False)
```
